## <u><strong> Data Vizualization </strong></u>

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

%matplotlib inline 

sns.set_style('darkgrid')
```

```python
for i in df.columns:
    if i=='bodyPart':
        pass
    else:
        plt.figure(figsize=(10,6))
        df.groupby(i).apply(lambda df:df['bodyPart'].value_counts().sum()).sort_values(ascending=False).plot.bar()
        plt.title(f'Number of {i} of Body part')
        plt.ylabel(f'Number of {i}')
```

```python
for i in df.columns:
    if i=='equipment':
        pass
    else:
        plt.figure(figsize=(10,6))
        df.groupby(i).apply(lambda df:df['equipment'].value_counts().sum()).sort_values(ascending=False).plot.bar()
        plt.title(f'Number of {i} of Equipment')
        plt.ylabel(f'Number of {i}')
```

```python
for i in df.columns:
    if i=='Exercise_name':
        pass
    else:
        plt.figure(figsize=(10,6))
        df.groupby(i).apply(lambda df:df['Exercise_name'].value_counts().sum()).sort_values(ascending=False).head(10).plot.bar()
        plt.title(f'Number of {i} of Exercise_name')
        plt.ylabel(f'Number of {i}')
```
